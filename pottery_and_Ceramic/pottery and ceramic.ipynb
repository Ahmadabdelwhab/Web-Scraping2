```python
import pandas as pd
```

```python
ceramic_yonobi1=pd.read_csv("ceramic yonobi1.csv")
ceramic_yonobi2=pd.read_csv("ceramic yonobi2.csv")
ceramic_yonobi3=pd.read_csv("ceramic yonobi3.csv")
ceramic_yonobi4=pd.read_csv("ceramic yonobi4.csv")
ceramic=pd.read_csv("ceramic.csv",encoding='unicode_escape')
ceramic2=pd.read_csv("ceramic2.csv")
ceramic3_accessories=pd.read_csv("ceramic3 accessories.csv")
ceramic3_bowls=pd.read_csv("ceramic3 bowls.csv")
ceramic3_mugs=pd.read_csv("ceramic3 mugs.csv")
ceramic3_plates=pd.read_csv("ceramic3 plates.csv")
ceramic3=pd.read_csv("ceramic3.csv")
ceramic_cups_enjoy=pd.read_csv("ceramic cups enjoy.csv")
```

```python
pottery_ceramic=pd.concat([ceramic_yonobi1,ceramic_yonobi2,ceramic_yonobi3,ceramic_yonobi4,ceramic,ceramic2,ceramic3_accessories,ceramic3_bowls,ceramic3_mugs,ceramic3_plates,ceramic3,ceramic_cups_enjoy])
```
