```python
import pandas as pd
```

```python
soy_candles_decorative=pd.read_csv("soy candles_decorative.csv",encoding='unicode_escape')
soy_candles_decorative2=pd.read_csv("soy candles_decorative2.csv",encoding='unicode_escape')
soy_candles=pd.read_csv("soy candles.csv",encoding='unicode_escape')
soy_candles2=pd.read_csv("soy candles2.csv",encoding='unicode_escape')
beeswax_candles1=pd.read_csv("beeswax candles1.csv",encoding='unicode_escape')
beeswax_candles3=pd.read_csv("beeswax candles3.csv",encoding='unicode_escape')
```

```python
decorative_df=pd.concat([soy_candles_decorative,soy_candles_decorative2], ignore_index=True)
```

```python
decorative_df['Caregory']='candles'
decorative_df['Subcaregory']='soy candles-decorative'
```

```python
decorative_df
```

```python
soy_candles_jars=pd.concat([soy_candles,soy_candles2], ignore_index=True)
```

```python
soy_candles_jars['Caregory']='candles'
soy_candles_jars['Subcaregory']='soy candles-jars'
```

```python
soy_candles_jars
```

```python
soy_candles=pd.concat([decorative_df,soy_candles_jars],  ignore_index=True)
```

```python
soy_candles
```

```python
beeswax_candles=pd.concat([beeswax_candles1,beeswax_candles3],ignore_index=True)
```

```python
beeswax_candles
```

```python
beeswax_candles['Caregory']='candles'
beeswax_candles['Subcaregory']='beeswax candles'
```

```python
beeswax_candles
```

```python
Candles=pd.concat([soy_candles,beeswax_candles],ignore_index=True)
```

```python
Candles
```

```python
Candles.to_csv("Candles dataset.csv",index=False)
```
